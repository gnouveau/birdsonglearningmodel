```python
# Fixed plots
%matplotlib inline
# Dynamic plots
# %matplotlib notebook

import matplotlib.pyplot as plt
from scipy.io import wavfile
import numpy as np

from IPython.display import Audio, display
```

```python
filename = 'simple'
```

```python
sr, song = wavfile.read('../data/{}.wav'.format(filename))

print("song length:", len(song))

song = song.astype(np.int32)
song = 2 * (song - np.min(song)) / (np.max(song) - np.min(song)) - 1
song = song - np.mean(song)
```

```python
sr, synth = wavfile.read('../data/{}_out.wav'.format(filename))

print("synth length:", len(synth))
```

```python
display(Audio(song, rate=sr))
display(Audio(synth, rate=sr))
```

```python
x = np.arange(len(song))
x = x / 44100

plt.figure(figsize=(10, 5))
plt.plot(x, song)
plt.title("song")
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(x, synth)
plt.title("synth")
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(x, song)
plt.plot(x, synth, alpha=0.5)
plt.title("comparison song / synth")
plt.show()
```
