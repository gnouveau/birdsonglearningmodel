Tests and analysis of the boari_synth_song_error code (boari_synth_song_error and carac_to_calculate_err_of_synth. cf utils.py in analysis folder, the GridAnalyser class and the learning_curve function)

Objective: Get an idea of the definition of the threshold to get only the syllables

```python
import matplotlib.pyplot as plt
from os.path import basename
import numpy as np
from scipy.io import wavfile
import sys

import birdsonganalysis as bsa
import utils

sys.path.append('../model')
from measures import bsa_measure, normalize_and_center
```

```python
# songs = ['ba_example', 'bells', 'bells_short', 'boari_short',
#          'flashcam', 'samba', 'simple']

songs = ['boari_short']

l_coefs = [{"fm": 0,
            "am": 0,
            "entropy": 1,
            "goodness": 0,
            "amplitude": 1,
            "pitch": 0},
           {"fm": 1,
            "am": 1,
            "entropy": 1,
            "goodness": 1,
            "amplitude": 1,
            "pitch": 1},
           {"fm": 1,
            "am": 1,
            "entropy": 5,
            "goodness": 1,
            "amplitude": 5,
            "pitch": 1}]

# l_coefs = [{"fm": 0,
#             "am": 0,
#             "entropy": 1,
#             "goodness": 0,
#             "amplitude": 1,
#             "pitch": 0}]

for song in songs:
    res1 = '{}:'.format(song)
    res2 = '{} whitout crop:'.format(song)
    res3 = '{} whitout threshold'.format(song)
    
    sr, tutor = wavfile.read('../data/{}.wav'.format(song))
    sr, synth = wavfile.read('../data/{}_out.wav'.format(song))
    
    tutor = normalize_and_center(tutor)
    tutor_feat = bsa.all_song_features(tutor, 44100,
                                       freq_range=256,
                                       fft_step=40,
                                       fft_size=1024)
    
    amp = bsa.song_amplitude(synth, 256, 40, 1024)
    amp = bsa.rescaling_one_feature(tutor_feat['amplitude'], amp)
    
    # Calculation without CROP
    # ******
    sort_amp = np.sort(amp)
    plt.figure(figsize=(10, 5))
    plt.plot(sort_amp)
    plt.title('sort_amp W/O CROP ({})'.format(song))
    plt.show()
    
    i_max_diff = np.argmax(utils._running_mean(np.diff(sort_amp),
                                               100))
    # th: threshold without the crop
    th = sort_amp[i_max_diff]
    # ******

    sort_amp = sort_amp[len(sort_amp)//10:] # CROP
    plt.figure(figsize=(10, 5))
    plt.plot(sort_amp)
    plt.title('sort_amp ({})'.format(song))
    plt.show()

    # Detail of i_max_diff calculations
    # ******
    amp_diff = np.diff(sort_amp)
    plt.figure(figsize=(10, 5))
    plt.plot(amp_diff)
    plt.title('amp_diff ({})'.format(song))
    plt.show()

    amp_diff_smoothed = utils._running_mean(np.diff(sort_amp),
                                            100)
    plt.figure(figsize=(10, 5))
    plt.plot(amp_diff_smoothed)
    plt.title('amp_diff_smoothed ({})'.format(song))
    plt.show()

    i_max_diff = np.argmax(utils._running_mean(np.diff(sort_amp),
                                               100))
    # ******
    
    threshold = sort_amp[i_max_diff]

    plt.figure(figsize=(10, 5))
    plt.plot(amp)
    plt.axhline(y=threshold, color="k", linewidth=1)
    plt.axhline(y=th, color="r", linewidth=1)
    plt.title('amp ({})'.format(song))
    plt.show()
    
    print("threshold W/O CROP ({})".format(song), th)
    print("threshold ({})".format(song), threshold)
    
    for coefs in l_coefs:
        msynth = bsa_measure(synth, 44100,
                             coefs=coefs, 
                             tutor_feat=tutor_feat)
        mtutor = bsa_measure(tutor, 44100, 
                             coefs=coefs, 
                             tutor_feat=tutor_feat)
        
        score = np.linalg.norm(msynth[amp > threshold] - mtutor[amp > threshold]) / np.sum(amp > threshold) * len(amp)
        res1 += '\t' + str(score)

        #WHITHOUT CROP
        score = np.linalg.norm(msynth[amp > th] - mtutor[amp > th]) / np.sum(amp > th) * len(amp)
        res2 += '\t' + str(score)

        # WITHOUT THRESHOLD
        score = np.linalg.norm(msynth - mtutor)
        res3 += '\t' + str(score)
    
    print(res1)
    print(res2)
    print(res3)
    print()
```
